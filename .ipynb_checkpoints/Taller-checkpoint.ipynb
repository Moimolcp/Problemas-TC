**3)** Sea el AFD $ M $  definido como $M = \{Q,\Sigma,\delta,q_0,F \}$ se definira $M'$ como el AFD que reconocera solamente las palabras que el AFD $M$ no reconoce.

$M' = \{Q',\Sigma',\delta',q_0',F' \}$

$Q' = Q \bigcup \{q_x\}$

$\Sigma' = \Sigma$

$q_0' = q_0$

$F' = (Q - F) \bigcup \{q_x\}$

Se definira $\delta'(q,a)$ para $q \in Q'$ y $a \in \Sigma'$ 

$\\ \delta'(q, a) =
\left\{
    \begin{array}{ll}
      q_x \quad si \quad q = q_x  \\
      q_x \quad si\quad\delta(q, a) = \emptyset  \\
      \delta(q, a) \quad si\quad\delta(q, a)\neq \emptyset  \\      
    \end{array}
\right.$

Se puede apreciar que el AFD $M'$ solo llegara a un estado final con palabras que $M$ no acepte.

**6)** Definir un AFD sobre el alfabeto $\{1,0\}$ que reconosca los numeros divisibles por 5.

el AFD $A$ recibira cadenas de bits iniciando por el mas significativo.

Supongamos tenemos  $X$ el cual es una cadena de bits que representa un numero decimal, si concadenaramos $X$ con la cadena de un solo 0 teniendo $X\cdot0$, podemos deducir el comportamiento en su representacion decimal

$X_{decimal} = X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n $ 

al concatenarle la cadena $0$ se tendria que su representacion decimal cambiaria a


$(X\cdot0)_{decimal} = (0)2^0 + X_0 2^1 + X_1 2^2 \cdot\cdot\cdot\cdot + X_n 2^{n+1} $


$(X\cdot0)_{decimal} = (0)2^0 + 2(X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n) $


$(X\cdot0)_{decimal} = 2(X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n) $

Podemos afirmar que al concadenarle un 0 a una cadena $X$ su represetacion en decimal es el doble, ahora si le concadenaramos un 1

$X_{decimal} = X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n $ 

al concatenarle la cadena $0$ se tendria que su representacion decimal cambiaria a


$(X\cdot1)_{decimal} = (1)2^0 + X_0 2^1 + X_1 2^2 \cdot\cdot\cdot\cdot + X_n 2^{n+1} $


$(X\cdot1)_{decimal} = (1)2^0 + 2(X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n) $


$(X\cdot1)_{decimal} = 1 + 2(X_0 2^0 + X_1 2^1 + X_2 2^2 \cdot\cdot\cdot\cdot + X_n 2^n) $

Podemos afirmar que al concadenarle un 1 a una cadena $X$ su represetacion en decimal es el doble mas 1.


Para que un numero $a$ sea divisible entre 5 se debe cumplir que $a \equiv 0 \,mod \,5$, lo que es equivalente a decir que se puede definir como $a = 5k$ siendo $k \in \mathbb{Z}$, todo numero se puede definir de la forma $a = 5k + r$ siendo $k \in \mathbb{Z}$ y siendo $r \in [0,4]$, Teniendo esto en cuenta podemos definir a $X_{decimal}$ de 5 maneras distintas

$X_{decimal} = 5k$

$X_{decimal} = 5k+1$

$X_{decimal} = 5k+2$

$X_{decimal} = 5k+3$

$X_{decimal} = 5k+4$

siendo $k \in \mathbb{Z}$, luego sabemos como se comporta $X$ si concadenaramos un 0 o un 1.


$(X\cdot0)_{decimal} = 2(X_{decimal}) = 2(5k) = 5l$   // Multiplo de 5

$(X\cdot0)_{decimal} = 2(X_{decimal}) = 2(5k +1) = 5l + 2$

$(X\cdot0)_{decimal} = 2(X_{decimal}) = 2(5k +2) = 5l + 4$

$(X\cdot0)_{decimal} = 2(X_{decimal}) = 2(5k +3)= 5l + 6 = 5j + 1$

$(X\cdot0)_{decimal} = 2(X_{decimal}) = 2(5k +4) = 5l + 8 = 5j + 3$  


$(X\cdot1)_{decimal} = 2(X_{decimal}) + 1 = 2(5k) +1 = 5l + 1 $ 

$(X\cdot1)_{decimal} = 2(X_{decimal}) + 1 = 2(5k +1)+ 1  = 5l + 3$

$(X\cdot1)_{decimal} = 2(X_{decimal}) + 1 = 2(5k +2)+ 1  = 5l $   // Multiplo de 5

$(X\cdot1)_{decimal} = 2(X_{decimal}) + 1 = 2(5k +3)+ 1 = 5l + 7 = 5j +2$

$(X\cdot1)_{decimal} = 2(X_{decimal}) + 1 = 2(5k +4)+ 1  = 5l + 9 = 5j + 4$ 

Con esto en cuenta podemos definir los estados de nuestro AFD $A$ como uno para cada residuo posible, de 0 a 4, definir las transciciones entre estados como arriba se menciona.

$ Q = \{ q_0,q_1,q_2,q_3,q_4\}$

$ \Sigma = \{ 0,1\}$

$ q_0 = q_0$

$ F = \{q_0\}$

$ \delta =$

|  | $0$ | $1$ |
|-|-|-|
| ${ q }_{ 0 }$ |${ q }_{ 0 }$|${ q }_{ 1 }$|
| ${ q }_{ 1 }$ |${ q }_{ 2 }$|${ q }_{ 3 }$|
| ${ q }_{ 2 }$ |${ q }_{ 4 }$|${ q }_{ 0 }$|
| ${ q }_{ 3 }$ |${ q }_{ 1 }$|${ q }_{ 2 }$|
| ${ q }_{ 4 }$ |${ q }_{ 3 }$|${ q }_{ 4 }$|
